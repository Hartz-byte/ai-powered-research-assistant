```python
# Imports
import os
import numpy as np
import torch
import torch.nn as nn
import pickle
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Data, Model, and Dicts
DATA_DIR = '../../data/ner/'
Xw_val = np.load(os.path.join(DATA_DIR, 'Xw_val.npy'))
Xp_val = np.load(os.path.join(DATA_DIR, 'Xp_val.npy'))
Yt_val = np.load(os.path.join(DATA_DIR, 'Yt_val.npy'))

with open(os.path.join(DATA_DIR, 'word2idx.pkl'), 'rb') as f:
    word2idx = pickle.load(f)
with open(os.path.join(DATA_DIR, 'pos2idx.pkl'), 'rb') as f:
    pos2idx = pickle.load(f)
with open(os.path.join(DATA_DIR, 'tag2idx.pkl'), 'rb') as f:
    tag2idx = pickle.load(f)
idx2tag = {i: t for t, i in tag2idx.items()}

VOCAB_SIZE = len(word2idx)
POS_SIZE = len(pos2idx)
TAG_SIZE = len(tag2idx)
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
class BiLSTM_NER(nn.Module):
    def __init__(self, vocab_size, pos_size, tag_size, emb_dim=100, pos_emb_dim=16, lstm_units=128):
        super().__init__()
        self.word_emb = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        self.pos_emb = nn.Embedding(pos_size, pos_emb_dim, padding_idx=0)
        self.lstm = nn.LSTM(emb_dim+pos_emb_dim, lstm_units, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(lstm_units*2, tag_size)
    def forward(self, Xw, Xp):
        w = self.word_emb(Xw)
        p = self.pos_emb(Xp)
        feats = torch.cat([w, p], dim=-1)
        lstm_out, _ = self.lstm(feats)
        out = self.fc(lstm_out)
        return out

# Load Model
SAVE_DIR = '../../models/ner'

model = BiLSTM_NER(VOCAB_SIZE, POS_SIZE, TAG_SIZE).to(DEVICE)
model.load_state_dict(torch.load(os.path.join(SAVE_DIR, "best_ner_model.pt"), map_location=DEVICE))
model.eval()
```

```python
Xw = torch.LongTensor(Xw_val).to(DEVICE)
Xp = torch.LongTensor(Xp_val).to(DEVICE)

with torch.no_grad():
    logits = model(Xw, Xp)
    preds = torch.argmax(logits, -1).cpu().numpy()

# Only consider non-padding positions
y_true, y_pred = [], []
for true_seq, pred_seq in zip(Yt_val, preds):
    for t, p in zip(true_seq, pred_seq):
        if t != 0:
            y_true.append(t)
            y_pred.append(p)

y_true_tag = [idx2tag[t] for t in y_true]
y_pred_tag = [idx2tag[p] for p in y_pred]

# Classification Report
print(classification_report(y_true_tag, y_pred_tag, labels=[t for t in tag2idx if t != '']))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_true_tag, y_pred_tag, labels=[t for t in tag2idx if t != ''])
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=False, fmt="d", 
            xticklabels=[t for t in tag2idx if t != ''], 
            yticklabels=[t for t in tag2idx if t != ''])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
idx2word = {i: w for w, i in word2idx.items()}

N = 5

for i in range(N):
    word_indices = Xw_val[i]
    true_tags = Yt_val[i]
    pred_tags = preds[i]
    
    words = [idx2word[idx] for idx in word_indices if idx != 0]
    true_labels = [idx2tag[t] for t in true_tags if t != 0]
    pred_labels = [idx2tag[p] for p in pred_tags if p != 0]
    
    print(f"Example {i+1}:")
    print("{:15} {:15} {:15}".format("Word", "True Tag", "Pred Tag"))
    print("-"*45)
    for w, t, p in zip(words, true_labels, pred_labels):
        print(f"{w:15} {t:15} {p:15}")
    print("\n")
```
