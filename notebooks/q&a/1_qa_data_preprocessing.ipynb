```python
# Imports
import json
import pandas as pd
import re
import pickle
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Load SQuAD datasets
def load_squad_data(train_file, dev_file):    
    with open(train_file, 'r', encoding='utf-8') as f:
        train_data = json.load(f)
    
    with open(dev_file, 'r', encoding='utf-8') as f:
        dev_data = json.load(f)
    
    print(f"Train version: {train_data['version']}")
    print(f"Dev version: {dev_data['version']}")
    
    return train_data, dev_data
```

```python
# Parse SQuAD data into structured format
def parse_squad_data(data):
    """Convert SQuAD JSON format to structured DataFrame"""
    
    examples = []
    
    for article in data['data']:
        title = article['title']
        
        for paragraph in article['paragraphs']:
            context = paragraph['context']
            
            for qa in paragraph['qas']:
                question_id = qa['id']
                question = qa['question']
                
                # Handle multiple answers (take first one for training)
                if qa['answers']:
                    answer = qa['answers'][0]
                    answer_text = answer['text']
                    answer_start = answer['answer_start']
                    answer_end = answer_start + len(answer_text)
                    is_impossible = False
                else:
                    # SQuAD 2.0 style impossible questions
                    answer_text = ""
                    answer_start = -1
                    answer_end = -1
                    is_impossible = True
                
                examples.append({
                    'id': question_id,
                    'title': title,
                    'context': context,
                    'question': question,
                    'answer_text': answer_text,
                    'answer_start': answer_start,
                    'answer_end': answer_end,
                    'is_impossible': is_impossible
                })
    
    return pd.DataFrame(examples)
```

```python
# Load and parse data
train_data, dev_data = load_squad_data('../../data/q&a/train-v1.1.json', 
                                       '../../data/q&a/dev-v1.1.json')

train_df = parse_squad_data(train_data)
dev_df = parse_squad_data(dev_data)

print(f"Train examples: {len(train_df)}")
print(f"Dev examples: {len(dev_df)}")
```

```python
# Data Analysis and Visualization
def analyze_qa_data(df, split_name):
    """Analyze Q&A dataset statistics"""
    
    print(f"\n=== {split_name} Dataset Analysis ===")
    
    # Text length statistics
    df['context_len'] = df['context'].str.split().apply(len)
    df['question_len'] = df['question'].str.split().apply(len)
    df['answer_len'] = df['answer_text'].str.split().apply(len)
    
    print(f"Context length - Mean: {df['context_len'].mean():.1f}, Max: {df['context_len'].max()}")
    print(f"Question length - Mean: {df['question_len'].mean():.1f}, Max: {df['question_len'].max()}")
    print(f"Answer length - Mean: {df['answer_len'].mean():.1f}, Max: {df['answer_len'].max()}")
    
    # Plot distributions
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    axes[0,0].hist(df['context_len'], bins=50, alpha=0.7, color='blue')
    axes[0,0].set_title(f'{split_name} - Context Length Distribution')
    axes[0,0].set_xlabel('Number of Words')
    
    axes[0,1].hist(df['question_len'], bins=30, alpha=0.7, color='green')
    axes[0,1].set_title(f'{split_name} - Question Length Distribution')
    axes[0,1].set_xlabel('Number of Words')
    
    axes[1,0].hist(df['answer_len'], bins=20, alpha=0.7, color='red')
    axes[1,0].set_title(f'{split_name} - Answer Length Distribution')
    axes[1,0].set_xlabel('Number of Words')
    
    # Answer position analysis
    df['answer_position_ratio'] = df['answer_start'] / df['context'].str.len()
    axes[1,1].hist(df['answer_position_ratio'], bins=30, alpha=0.7, color='orange')
    axes[1,1].set_title(f'{split_name} - Answer Position in Context')
    axes[1,1].set_xlabel('Position Ratio (0=start, 1=end)')
    
    plt.tight_layout()
    plt.show()
    
    return df
```

```python
# Analyze both datasets
train_df = analyze_qa_data(train_df, "Training")
dev_df = analyze_qa_data(dev_df, "Development")
```

```python
# Text Cleaning and Tokenization
def clean_text(text):
    """Basic text cleaning"""
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()
    return text

def simple_tokenize(text):
    """Simple word tokenization"""
    text = re.sub(r"([.!?])", r" \1 ", text)
    text = re.sub(r"[^a-zA-Z0-9.!?]+", r" ", text)
    return text.split()

# Build vocabulary
def build_vocab(train_df, dev_df, min_freq=3, max_vocab_size=50000):
    """Build vocabulary from questions and contexts"""
    
    word_counts = Counter()
    
    # Count words in contexts and questions
    for df in [train_df, dev_df]:
        for text in df['context']:
            word_counts.update(simple_tokenize(clean_text(text.lower())))
        for text in df['question']:
            word_counts.update(simple_tokenize(clean_text(text.lower())))
    
    # Special tokens
    special_tokens = ['<PAD>', '<UNK>', '<CLS>', '<SEP>']
    
    # Create vocabulary
    vocab = special_tokens.copy()
    
    # Add frequent words
    frequent_words = [word for word, count in word_counts.most_common() 
                     if count >= min_freq][:max_vocab_size - len(special_tokens)]
    
    vocab.extend(frequent_words)
    
    # Create word-to-index mapping
    word2idx = {word: idx for idx, word in enumerate(vocab)}
    idx2word = {idx: word for word, idx in word2idx.items()}
    
    print(f"Vocabulary size: {len(vocab)}")
    print(f"Most common words: {frequent_words[:10]}")
    
    return word2idx, idx2word
```

```python
# Character-level answer position mapping
def find_answer_positions(context, question, answer_text, answer_start):
    """Find token-level start and end positions for answers"""
    
    if answer_start == -1:  # Impossible question
        return -1, -1
    
    # Tokenize context
    context_tokens = simple_tokenize(clean_text(context.lower()))
    answer_tokens = simple_tokenize(clean_text(answer_text.lower()))
    
    context_text = clean_text(context.lower())
    answer_text_clean = clean_text(answer_text.lower())
    
    # Find character positions
    char_start = context_text.find(answer_text_clean)
    char_end = char_start + len(answer_text_clean)
    
    if char_start == -1:
        return -1, -1
    
    # Convert to token positions (approximate)
    tokens_before_start = len(simple_tokenize(context_text[:char_start]))
    tokens_in_answer = len(simple_tokenize(answer_text_clean))
    
    token_start = tokens_before_start
    token_end = token_start + tokens_in_answer - 1
    
    # Ensure positions are within bounds
    token_start = max(0, min(token_start, len(context_tokens) - 1))
    token_end = max(token_start, min(token_end, len(context_tokens) - 1))
    
    return token_start, token_end

# Process datasets with tokenization and position mapping
def process_qa_dataset(df, word2idx, max_context_len=400, max_question_len=50):
    """Process Q&A dataset with tokenization and encoding"""
    
    processed_examples = []
    
    for idx, row in df.iterrows():
        # Clean and tokenize
        context_clean = clean_text(row['context'].lower())
        question_clean = clean_text(row['question'].lower())
        
        context_tokens = simple_tokenize(context_clean)
        question_tokens = simple_tokenize(question_clean)
        
        # Find answer positions
        if not row['is_impossible']:
            start_pos, end_pos = find_answer_positions(
                row['context'], row['question'], 
                row['answer_text'], row['answer_start']
            )
        else:
            start_pos, end_pos = -1, -1
        
        # Encode tokens to indices
        context_ids = [word2idx.get(token, word2idx['<UNK>']) for token in context_tokens[:max_context_len]]
        question_ids = [word2idx.get(token, word2idx['<UNK>']) for token in question_tokens[:max_question_len]]
        
        # Pad sequences
        context_ids += [word2idx['<PAD>']] * (max_context_len - len(context_ids))
        question_ids += [word2idx['<PAD>']] * (max_question_len - len(question_ids))
        
        processed_examples.append({
            'id': row['id'],
            'context_ids': context_ids,
            'question_ids': question_ids,
            'context_tokens': context_tokens[:max_context_len],
            'question_tokens': question_tokens[:max_question_len],
            'start_position': start_pos,
            'end_position': end_pos,
            'is_impossible': row['is_impossible'],
            'answer_text': row['answer_text']
        })
        
        if idx % 10000 == 0:
            print(f"Processed {idx} examples...")
    
    return processed_examples
```

```python
# Build vocabulary and process datasets
word2idx, idx2word = build_vocab(train_df, dev_df)

print("Processing training data...")
train_processed = process_qa_dataset(train_df, word2idx)

print("Processing development data...")
dev_processed = process_qa_dataset(dev_df, word2idx)
```

```python
# Save vocabulary
with open('../../data/q&a/vocab.pkl', 'wb') as f:
    pickle.dump({'word2idx': word2idx, 'idx2word': idx2word}, f)

# Save processed datasets
with open('../../data/q&a/train_processed.pkl', 'wb') as f:
    pickle.dump(train_processed, f)

with open('../../data/q&a/dev_processed.pkl', 'wb') as f:
    pickle.dump(dev_processed, f)

print("Preprocessing completed and saved!")
print(f"Processed {len(train_processed)} training examples")
print(f"Processed {len(dev_processed)} development examples")
```
