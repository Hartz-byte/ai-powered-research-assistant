```python
# Imports
import json
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load vocab
vocab_path = '../../models/sentiment/vocab.json'

with open(vocab_path, 'r') as f:
    vocab = json.load(f)
vocab_size = len(vocab)
```

```python
# Model definition 
class BiLSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim=256, hidden_dim=128, output_dim=3, num_layers=2, dropout=0.3):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=vocab['<PAD>'])
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers, batch_first=True, dropout=dropout, bidirectional=True)
        self.fc = nn.Linear(hidden_dim * 2, output_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        embedded = self.embedding(x)
        lstm_out, (hidden, _) = self.lstm(embedded)
        hidden_cat = torch.cat((hidden[-2], hidden[-1]), dim=1)
        out = self.dropout(hidden_cat)
        return self.fc(out)

# Model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BiLSTMClassifier(vocab_size).to(device)
model.load_state_dict(torch.load('../../models/sentiment/best_sentiment_model.pt', map_location=device))
model.eval()
```

```python
# Preprocessing and Prediction Utilities
def preprocess_text(text, word2idx, max_len=100):
    import re
    text = str(text).lower()
    text = re.sub(r'https?://\\S+|www\\.\\S+', '', text)
    text = re.sub(r"[^a-z0-9\\s']", '', text)
    tokens = text.split()
    ids = [word2idx.get(t, word2idx['<UNK>']) for t in tokens]
    if len(ids) < max_len:
        ids.extend([word2idx['<PAD>']] * (max_len - len(ids)))
    else:
        ids = ids[:max_len]
    return torch.tensor([ids], dtype=torch.long).to(device)

def predict_sentiment(text, model, word2idx):
    model.eval()
    tensor = preprocess_text(text, word2idx)
    with torch.no_grad():
        logits = model(tensor)
        probs = F.softmax(logits, dim=1)
        class_id = probs.argmax(dim=1).item()
        class_map_reverse = {0: 'negative', 1: 'neutral', 2: 'positive'}
        return class_map_reverse[class_id], float(probs[0, class_id])
```

```python
# Load validation set
val_df = pd.read_csv('../../data/sentiment-analysis/processed_val.csv')
all_preds = []
all_true = []

for text, label in zip(val_df['clean_text'], val_df['label']):
    pred_class, _ = predict_sentiment(text, model, vocab)
    str2label = {'negative': 0, 'neutral': 1, 'positive': 2}
    all_preds.append(str2label[pred_class])
    all_true.append(label)

# Converting to numpy arrays for sklearn metrics
all_true = np.array(all_true)
all_preds = np.array(all_preds)
```

```python
# Overall metrics
acc = accuracy_score(all_true, all_preds)
f1 = f1_score(all_true, all_preds, average='weighted')
print(f"Validation Accuracy: {acc:.3f}")
print(f"Validation F1 Score: {f1:.3f}")
print("\nClassification Report:")
print(classification_report(all_true, all_preds, target_names=['negative', 'neutral', 'positive']))

# Confusion Matrix
cm = confusion_matrix(all_true, all_preds)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues',
            xticklabels=['negative', 'neutral', 'positive'],
            yticklabels=['negative', 'neutral', 'positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Try inference on new text
example = "Absolutely loved the fast customer service!"
sentiment, conf = predict_sentiment(example, model, vocab)
print(f'Text: {example}')
print(f'Predicted Sentiment: {sentiment} (confidence: {conf:.2f})')
```
