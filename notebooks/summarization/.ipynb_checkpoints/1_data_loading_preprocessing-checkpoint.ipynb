```python
# Imports
import pandas as pd
import re
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Load data
processed_path = '../../data/summarization/processed_train.csv'
train_path = '../../data/summarization/processed_train_split.csv'
val_path = '../../data/summarization/processed_val_split.csv'
data_path = '../../data/summarization/train.csv'
df = pd.read_csv(data_path)

# Sample ~57,000 rows randomly
df = df.sample(n=57000, random_state=42).reset_index(drop=True)

# Inspect a sample
print(df[['article', 'highlights']].iloc[0])
```

```python
# Clean text functions
def clean_text(text):
    text = str(text).replace('\n', ' ').lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r"[^a-z0-9\s.,']", " ", text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_article'] = df['article'].apply(clean_text)
df['clean_summary'] = df['highlights'].apply(clean_text)
```

```python
# Remove empty or malformed samples
df = df[(df['clean_article'].str.len() > 0) & (df['clean_summary'].str.len() > 0)]

print(f"Total samples after sampling & cleaning: {len(df)}")
```

```python
# Length distributions
df['article_len'] = df['clean_article'].str.split().apply(len)
df['summary_len'] = df['clean_summary'].str.split().apply(len)

print("Article length stats:")
print(df['article_len'].describe())
print("\nSummary length stats:")
print(df['summary_len'].describe())
```

```python
# Plot length distributions for debugging
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(df['article_len'], bins=50, color='cornflowerblue')
plt.title('Article Length Distribution')
plt.xlabel('Tokens')
plt.ylabel('Count')

plt.subplot(1,2,2)
plt.hist(df['summary_len'], bins=25, color='coral')
plt.title('Summary Length Distribution')
plt.xlabel('Tokens')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Print label balancing stats
print(f"Total samples: {len(df)}")
print(f"Min article length: {df['article_len'].min()} | Max: {df['article_len'].max()}")
print(f"Min summary length: {df['summary_len'].min()} | Max: {df['summary_len'].max()}")
```

```python
# Save clean, standardized splits
train_df, val_df = train_test_split(
    df[['id', 'clean_article', 'clean_summary']],
    test_size=0.15, random_state=42
)
train_df.to_csv('../../data/summarization/processed_train.csv', index=False)
val_df.to_csv('../../data/summarization/processed_val.csv', index=False)

print(f"Train samples: {len(train_df)}, Val samples: {len(val_df)}")
```
